
### ==> Data Transformation
***

**Create occupancy percentage column**

```python
df_agg_booking.head(4)
```

```python
df_agg_booking.loc[:,"occ_pct"] = df_agg_booking["successful_bookings"]/ df_agg_booking["capacity"]*100
```

```python
df_agg_booking.head(4)
```

**Using .loc[:,] approach to get rid of SettingWithCopyWarning**

```python
df_agg_booking.loc[:,'occ_pct'] = df_agg_booking['occ_pct'].apply(lambda x: round(x, 2))
df_agg_booking.head(3)
```

```python
df_bookings.head()
```

**When merging two DataFrames in pandas that have columns with the same name, the suffixes parameter is used to differentiate the overlapping column names in the resulting merged DataFrame. By specifying suffixes, we can avoid column name collisions and make it clear which DataFrame each column came from.**

```python
df= pd.merge(df,df_hotels, on="property_id",suffixes=('_df', '_df_hotels'))
df.head(3)
```

**Now droping the duplicate columns** 

```python
df_cleaned = df.drop(columns=[
    'property_name_y', 'category_y', 'city_y',
    'property_name_df', 'category_df', 'city_df',
    'property_name_df_hotels', 'category_df_hotels', 'city_df_hotels'
])
```
